```python
import os
from glob import glob 
import shutil
import tqdm

from processing.vnmese_utils import normalize_VNmese, remove_diacritics
```

```python
htdemucs_path = 'out/htdemucs_preprocessed_100'
vocal_files = glob(f'{htdemucs_path}/*/vocals.mp3') 

for vocal_file in tqdm.tqdm(vocal_files):
    shutil.copyfile(vocal_file, f'{htdemucs_path}/{vocal_file.split("/")[-2]}.mp3')


```

```python
preproc_path = 'out/quoc_master_1st_1000_preprocessed/quoc_master_1st_1000_preprocessed'

preproc_songs = os.listdir(f'{preproc_path}')
# preproc_songs = [preproc_songs[i].split('.')[0] for i in range(len(preproc_songs))]
print(preproc_songs)
```

```python
test_path = './out/master_1st_100_preprocessed' 
test_song_paths = glob(f'{test_path}/*.origin.lrc')
test_songnames = [test_song_paths[i].split('/')[-1].split('.origin.lrc')[0] for i in range(len(test_song_paths))]
print(test_song_paths)
print(test_songnames) 
```

```python
output_path = 'out/quoc_master_1st_100_preprocessed'
import tqdm

preproc_songnames = [songname.split('.mp3')[0].split('.origin.lrc')[0] for songname in preproc_songs]
deViet_demuc_songs = [remove_diacritics(songname) for songname in preproc_songnames]

for songname in tqdm.tqdm(test_songnames):
    deViet_songname = remove_diacritics(songname)
    deViet_demuc_id = deViet_demuc_songs.index(deViet_songname)

    shutil.copyfile(f'{preproc_path}/{preproc_songnames[deViet_demuc_id]}.mp3', f'{output_path}/{preproc_songnames[deViet_demuc_id]}.mp3')
    shutil.copyfile(f'{preproc_path}/{preproc_songnames[deViet_demuc_id]}.origin.lrc', f'{output_path}/{preproc_songnames[deViet_demuc_id]}.origin.lrc')
```
