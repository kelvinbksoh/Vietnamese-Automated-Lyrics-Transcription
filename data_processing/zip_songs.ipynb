```python
import zipfile
import os
from glob import glob
import tqdm 
def compress_files(file_list, output_zip):
    # Create a new zip file
    with zipfile.ZipFile(output_zip, 'w') as zipf:
        for file in tqdm.tqdm(file_list):
            if os.path.isfile(file):
                # Add file to the zip, maintaining the directory structure
                zipf.write(file, os.path.basename(file))
            else:
                print(f"File {file} does not exist")
```

```python
songs = glob('/home/anh/Documents/vietnamese-song-scraping/out/master_1st_17000/*.mp3')
print(len(songs))
songs_with_lyrics = []
for song in songs:
    lyric_file = song.replace('.mp3', '.origin.lrc')
    if os.path.isfile(lyric_file):
        songs_with_lyrics.append(song)
        songs_with_lyrics.append(lyric_file)
print(len(songs_with_lyrics))
```

```python
import pandas as pd
link_df = pd.read_csv('out/unique_songname_links.csv')
links = link_df['download_link'].to_list()
```

```python
tiny_links = [x.split('https://zingmp3.vn/bai-hat/')[1].split('/')[0].lower() for x in links]
print(tiny_links)
```

```python
import re
full_songs_with_lyrics = list(set([os.path.basename(x).split('.mp3')[0].split('.origin.lrc')[0] for x in songs_with_lyrics]))

songs_with_lyrics = list(set([os.path.basename(x).split('.mp3')[0].split('.origin.lrc')[0].lower().replace(')','').replace('(','') for x in songs_with_lyrics]))
print(songs_with_lyrics)
# cleaned_text = re.sub(r'[,!?-]', '', text)

# songs_with_lyrics = [re.sub(r'[,!?&:]', '', x) for x in songs_with_lyrics]

# songs_with_lyrics = [re.sub(r'[^a-zA-Z0-9\-_.~]', '', x) for x in songs_with_lyrics]
```

```python
songs_with_lyrics = [' '.join([y for y in x.split(' ') if y != '']) for x in songs_with_lyrics]
print(songs_with_lyrics)
```

```python
from url_scrape import remove_diacritics
```

```python
corresponding_links = []
none_links = []
dedia = []
for song in tqdm.tqdm(songs_with_lyrics):
    de_diacritic = remove_diacritics(song).strip()
    de_diacritic = de_diacritic.replace('...','-').replace(',', '-')
    de_diacritic = de_diacritic.replace('.','-').replace('', '-')
    de_diacritic = de_diacritic.replace('&','-').replace('', '-').replace('', '-')
    de_diacritic = de_diacritic.replace('','-').replace('\'','-').replace('', '-')

    de_diacritic = '-'.join([y for y in de_diacritic.split(' ') if y != ''])
    de_diacritic = re.sub(r'[^a-zA-Z0-9\-_.~]', '', de_diacritic)
    de_diacritic = '-'.join([y for y in de_diacritic.split('-') if y != ''])
    dedia.append(de_diacritic)
    for i, link in enumerate(tiny_links):
        if de_diacritic in link:
            corresponding_links.append(links[i])
            break
    else:
        corresponding_links.append('NONE FOUND')
        none_links.append(song + ' - ' + de_diacritic)
```

```python
df = pd.DataFrame()
df['song'] = full_songs_with_lyrics
df['link'] = corresponding_links
df['prefix'] = dedia
df.to_csv('out/8k_songs.csv', index=False)

```
